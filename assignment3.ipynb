In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
%matplotlib ipympl
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import scipy.stats
```

As Ferreira, Fisher, & Konig did in [their paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) we would like to know how certain we are about the mean of a data set is a given number. It is useful to calculate then a sampling distribution of our data to measure how the mean spreads for a given population.

In order to calculate the uncertainty of the sampling distribution, it would be expected to be as **normal** as possible, so remembering the hypotheses of the **Central limit theorem**, it is mandatory to have a large number of independent and identically distributed variables, to guarantee that the random variable $ \sqrt(n)(X_{m} - \mu) \to \mathcal{N}(0,\sigma^2)$ where $X_{m}$ is the mean of the sequence of random variables.

```python
df_means = pd.DataFrame()
```

```python
df_means[1992] = pd.Series([np.random.normal(32000,200000,3650).mean() for i in range(1000)])
df_means[1993] = pd.Series([np.random.normal(43000,100000,3650).mean() for i in range(1000)])
df_means[1994] = pd.Series([np.random.normal(43500,140000,3650).mean() for i in range(1000)])
df_means[1995] = pd.Series([np.random.normal(48000,70000,3650).mean() for i in range(1000)])
```

```python
std_serie = df_means.std()
mean_serie = df_means.mean()
```

```python
### Create the main 4 bars with the calculated statistical features
fig, ax = plt.subplots()
for i in range(4):
    main_graph = ax.bar(df_means.columns,mean_serie,width = 1,align = "center",edgecolor = "black",
            tick_label = df_means.columns,yerr = 1.96*std_serie,capsize=12)

### Set the plot beautiness
plt.rcParams["figure.figsize"] = (6,6)
ax.set_title('Hardest option assessment')
ax.set_yticks(np.linspace(-6263.0,51595.8,num = 10).round(decimals=1))
ax.spines['bottom'].set_position(('data', 0))
ax.spines[['top','right']].set_visible(False)

#-- Set default region
shadw = patches.Rectangle((0,6594), width=1996, height=13023-6594, alpha=0.3, edgecolor = 'black', facecolor='gray')
ax.add_patch(shadw)
#-- Set legends for the default region
x_zero = ax.get_xlim()[0]
top_text= ax.annotate(f'{shadw.get_y()+shadw.get_height()}',
                      (x_zero,shadw.get_y()+shadw.get_height()),alpha =0.7,
                      xytext = (-6,6),textcoords = "offset pixels",
                      bbox=dict(facecolor='gray', alpha=0.15),ha='right', va='bottom')
bottom_text= ax.annotate(f'{shadw.get_y()}',
                         (x_zero,shadw.get_y()),alpha =0.7,
                         xytext = (-6,-6),textcoords = "offset pixels",
                         bbox=dict(facecolor='gray', alpha=0.15),ha='right', va='top')

#-- Colorbar
# norm = mpl.colors.Normalize(vmin=0, vmax=1)
norm = mpl.colors.BoundaryNorm(np.linspace(0, 1, 10).round(decimals=2), mpl.cm.Reds.N)
sm = plt.cm.ScalarMappable(cmap=mpl.cm.Reds, norm=norm)
sm.set_array([])
plt.colorbar(sm, ticks=np.linspace(0, 1, 10).round(decimals=2),orientation='horizontal',
             drawedges = True,)

### Probability that the distribution represented by the bar is contained in the region
def p(mean,std,l_inf,l_sup):
    dis = scipy.stats.norm(mean, std)
    return dis.cdf(l_sup) - dis.cdf(l_inf)

#-- Color the main bars
cmap = mpl.cm.Reds
for i,bar in enumerate(main_graph):
    proba = p(mean_serie.iloc[i],std_serie.iloc[i],*shadw.xy)
    bar.set_facecolor(cmap(proba))

### Interactivity
class interacting_graph:
    def __init__(self,region,recs,stats,top,bottom):
        self.press = None
        self.region = region
        self.recs = recs
        self.stats = stats
        self.top = top
        self.bottom = bottom
    
    def connect(self):
        # Connect to all the events we need
        self.cidpress = self.region.figure.canvas.mpl_connect('button_press_event', self.on_press)
        self.cidrelease = self.region.figure.canvas.mpl_connect('button_release_event', self.on_release)
        self.cidmotion = self.region.figure.canvas.mpl_connect('motion_notify_event', self.on_motion)
        
    def on_press(self, event):
        # Check whether mouse clicked; if so, store some data.
        if event.inaxes != self.region.axes:
            return
        # contains, attrd = self.region.contains(event)
        # if not contains:
        #     return
        self.press = event.ydata
        
    def on_motion(self, event):
        # If the mouse is moving after a click, then change our shape.
        if self.press is None:
            return
        
        y0 = self.press
        if event.ydata > y0:
            # Redrawing shaded region
            dy = event.ydata - y0
            self.region.set_y(y0)
            self.region.set_height(dy)
            
            # Repositioning annotations
            self.top.xy = (x_zero,y0+dy)
            self.bottom.xy = (x_zero,y0)
            self.top.set_text(f'{(y0+dy).round(decimals=1)}'[:7])
            self.bottom.set_text(f'{y0.round(decimals=1)}'[:7])
            
            # Recoloring bars
            cmap = mpl.cm.Reds
            for i,bar in enumerate(self.recs):
                proba = p(self.stats[0].iloc[i],self.stats[1].iloc[i],y0,event.ydata)
                bar.set_facecolor(cmap(proba))
            
            self.region.figure.canvas.draw()
            
    def on_release(self, event):
        # Clear button press information.
        self.press = None
        self.region.figure.canvas.draw()

    def disconnect(self):
        # Disconnect all callbacks.
        self.region.figure.canvas.mpl_disconnect(self.cidpress)
        self.region.figure.canvas.mpl_disconnect(self.cidrelease)
        self.region.figure.canvas.mpl_disconnect(self.cidmotion)

gr = interacting_graph(shadw,main_graph,(mean_serie,std_serie),top_text,bottom_text)
gr.connect()
```
